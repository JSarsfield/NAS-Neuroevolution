```python
def gaussian(x, freq=0.314, amp=2, vshift=-1):
    return (np.sign(freq)*(amp*(math.e**(-(0.5*(x/freq)**2)))))+vshift
```

```python
def sin(x, freq=3.14, amp=1, vshift=0):
    return (amp*np.sin(freq*x))+vshift
```

```python
def exp(x, freq=1, amp=0):
    return (freq*x**2)+amp
```
